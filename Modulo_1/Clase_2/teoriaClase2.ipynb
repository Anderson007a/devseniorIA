# Condicionales y bucles

> condicionales -> if/elif/else, es una estructura de control, controla el flujo del programa


```python
nombre = "Jose"
promedio = 6.4

if promedio >= 7:
    print(f"Felicitaciones {nombre}")
else:
    print(f"Necesitas mejorar {nombre}")
```

```python
nombre = "Pepito"
promedio = 4

if promedio >= 7:
    print(f"Felicitaciones {nombre}")
elif promedio >= 5:
    print(f"{nombre} puedes mejorar")
else:
    print(f"Necesitas mejorar {nombre}")
```

# Bucles

> For, es una estructura de iteración, recorre todo el elemento


```python
learners = [
    {"nombre": "Diana", "promedio": 8.5},
    {"nombre": "Juan", "promedio": 4.5},
    {"nombre": "Alexander", "promedio": 5.1}
]

for learner in learners:
    if learner["promedio"] >= 7:
      print(f"Felicitaciones {learner['nombre']}")
    elif promedio >= 5:
        print(f"{learner['nombre']} puedes mejorar")
    else:
        print(f"Necesitas mejorar {learner['nombre']}")
```
