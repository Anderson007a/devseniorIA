# PARTE 1 — Resumen

---

## 1. SELECT — ¿Qué quiero ver?

Recupera columnas de una tabla. Es la instrucción más fundamental de SQL.

```sql
SELECT columna1, columna2   -- columnas específicas
SELECT *                    -- todas las columnas
FROM nombre_tabla
LIMIT N                     -- máximo N filas
```

**Ejemplo:**
```sql
SELECT nombre, ciudad, nivel
FROM candidatos
LIMIT 5
```

---

## 2. DISTINCT — Eliminar duplicados

Muestra solo valores únicos de una columna. Sin DISTINCT, cada fila aparece aunque se repita.

```sql
SELECT DISTINCT columna
FROM tabla
ORDER BY columna   -- opcional, ordena el resultado
```

**Ejemplo:**
```sql
SELECT DISTINCT ciudad
FROM candidatos
ORDER BY ciudad
```

---

## 3. Alias — Nombres legibles

Renombra columnas en el resultado. No modifica la tabla original.

```sql
SELECT columna AS "Nombre Legible"
FROM tabla
```

**Ejemplo:**
```sql
SELECT nombre AS "Candidato", salario_esperado_cop AS "Salario (COP)"
FROM candidatos
```

---

## 4. WHERE — Filtrar filas

Solo pasan las filas que cumplen la condición. Se evalúa **antes** de SELECT.

**Orden obligatorio en SQL:**
```
SELECT → FROM → WHERE → GROUP BY → HAVING → ORDER BY → LIMIT
```

| Operador | Qué hace | Ejemplo |
|----------|----------|---------|
| `=` | Igual exacto | `ciudad = 'Bogotá'` |
| `AND` | Ambas condiciones | `ciudad = 'Bogotá' AND disponible = 'Sí'` |
| `OR` | Cualquier condición | `nivel = 'Senior' OR nivel = 'Lead'` |
| `IN` | Está en la lista | `ciudad IN ('Bogotá', 'Medellín', 'Cali')` |
| `BETWEEN` | Rango inclusivo | `anos_experiencia BETWEEN 3 AND 6` |

> ⚠️ Los textos siempre entre **comillas simples** `'`. Los paréntesis importan con AND/OR.

---

## 5. GROUP BY + Funciones de agregación

Agrupa filas con el mismo valor y aplica una función que calcula un número resumen por grupo.

| Función | Qué calcula |
|---------|-------------|
| `COUNT(*)` | Cuántas filas hay en el grupo |
| `AVG(col)` | Promedio de la columna |
| `SUM(col)` | Suma de la columna |
| `MAX(col)` | Valor más alto |
| `MIN(col)` | Valor más bajo |
| `ROUND(val, n)` | Redondea a n decimales |

```sql
SELECT columna_grupo, COUNT(*) AS total
FROM tabla
GROUP BY columna_grupo
ORDER BY total DESC
```

---

## 6. HAVING — Filtrar grupos

Filtra **después** de agrupar. WHERE no puede usarse con funciones de agregación — HAVING sí.

```sql
SELECT ciudad, COUNT(*) AS total
FROM candidatos
GROUP BY ciudad
HAVING COUNT(*) > 50   -- solo ciudades con más de 50 candidatos
```

| | WHERE | HAVING |
|--|-------|--------|
| **Cuándo** | Antes de agrupar | Después de agrupar |
| **Filtra** | Filas individuales | Grupos completos |
| **Acepta COUNT(), AVG()...** | ❌ No | ✅ Sí |

---

## 7. JOIN — Cruzar tablas

Combina filas de dos tablas usando la columna que tienen en común (PK → FK).

```sql
FROM tabla_a a
INNER JOIN tabla_b b ON a.id_comun = b.id_comun
```

| Tipo | Qué devuelve |
|------|--------------|
| `INNER JOIN` | Solo filas con coincidencia en AMBAS tablas |
| `LEFT JOIN` | Todas las filas de la tabla izquierda, con o sin coincidencia |

**Ejemplo triple JOIN:**
```sql
FROM postulaciones p
INNER JOIN candidatos c ON p.candidato_id = c.candidato_id
INNER JOIN ofertas    o ON p.oferta_id    = o.oferta_id
```
